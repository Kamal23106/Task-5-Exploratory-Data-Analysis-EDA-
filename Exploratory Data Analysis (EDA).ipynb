```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Dataset = pd.read_csv(r"C:\Users\USER\Downloads\Titanic_Dataset.csv")
```

```python
print("Dataset Info:")
print(Dataset.info())

print("\nDataset Description:")
print(Dataset.describe())
```

```python
print("\nValue Counts for Categorical Columns:")
for col in Dataset.select_dtypes(include='object').columns:
    print(f"\nValue counts for {col}:")
    print(Dataset[col].value_counts())
```

```python
print("\nGenerating Pairplot...")
sns.pairplot(Dataset)
plt.show()
```

```python
print("\nGenerating Correlation Heatmap...")
plt.figure(figsize=(12, 8))
sns.heatmap(Dataset.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
print("\nGenerating Histograms...")
Dataset.hist(figsize=(15, 10), bins=20, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
print("\nGenerating Boxplots...")
numerical_cols = Dataset.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=Dataset[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
print("\nGenerating Scatterplots...")
if len(numerical_cols) >= 2:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=Dataset[numerical_cols[0]], y=Dataset[numerical_cols[1]])
    plt.title(f'Scatterplot between {numerical_cols[0]} and {numerical_cols[1]}')
    plt.xlabel(numerical_cols[0])
    plt.ylabel(numerical_cols[1])
    plt.show()
```

```python
# --- Observations ---

print("\n\n--- Observations for Each Visual ---\n")
print("Dataset has 891 rows and 12 columns. Some missing values exist in Age, Cabin, and Embarked.")
print("More males than females. Most passengers from Southampton. Majority traveled in 3rd class.")
print("Pairplot shows females and 1st-class passengers had higher survival.")
print("Heatmap: Pclass negatively correlates with Survived; Fare positively correlates.")
print("Histograms: Age is right-skewed, Fare highly skewed.")
print("Boxplots: Fare has major outliers. Age has mild outliers.")
print("Scatterplots: No strong relationship between Age and Fare.")
```

```python
# --- Summary of Findings ---

print("\n\n--- Final Summary of Findings ---\n")
print("- Missing values in Age, Cabin, Embarked.")
print("- Outliers detected in Fare.")
print("- Sex and Pclass are strong predictors of survival.")
print("- Age mostly between 20-40. Fare highly skewed.")
print("- Most passengers were 3rd class, embarked from Southampton.")
print("- Higher-class, female passengers had higher survival rates.")
```
